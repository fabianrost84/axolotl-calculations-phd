```python
cell_length_data_plot = cell_length_data.query('cell_population != "non-regenerating"').copy()
@sp.vectorize
def time_label(a):
    if a == 0:
        return 'Day 0'
    elif a == 3:
        return 'Day 3'
cell_length_data_plot['time'] = time_label(cell_length_data_plot['time'])
cell_length_data_plot.head()
```

```python
fig = plt.figure(figsize = (80.0 / 25.4, 45.0 / 25.4))
sns.set_style('whitegrid', 
              {'font.family': ['Helvetica LT Std'],
               'font.sans-serif':  ['Helvetica LT Std'],               
                                   })
sns.set_context('paper')
mpl.rcParams['xtick.major.pad'] = 15.0
g = sns.swarmplot(x='ID', y='cell_length', hue = 'time', data = cell_length_data_plot);
g.set_xticklabels(g.get_xticklabels(), rotation=45)
plt.xlabel('Animal ID')
plt.ylabel(u'Cell length (μm)')
plt.ylim(10, 21)
plt.legend(title = None, loc = 'upper right')
plt.show()
```
